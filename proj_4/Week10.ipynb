
## 載入所用package 及 91APP訂單資料



用mlxtend包來進行apriori分析


```python
import pandas as pd
import matplotlib as mlt
import numpy as np
from io import BytesIO as StringIO
from mlxtend.frequent_patterns  import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df_orders = pd.read_csv('../../Ntu_Orders.csv')
df_promotions = pd.read_csv('../../PromotionOrders.csv')
df_members = pd.read_csv('../../Member.csv')
```

## 顧客購物類別關聯性

製作每個顧客ID曾經購買的商品類別表

```python
df_test= df_orders.groupby(['MemberId', 'LevelFourCategoryName'])['Quantity'].sum().unstack(fill_value=0).astype('bool')
df_test.head()
```



進行分析 發現 T恤/帽T 與其他衣物的搭配經常出現


```python
df_apriori = apriori(df_test, min_support=0.07, use_colnames=True)
df_conf = association_rules(df_apriori, metric="lift", min_threshold=1)
df_conf=df_conf[df_conf["confidence"]>0.6]
df_conf
```



對較高層級Levelthree的商品類別作分析 不過並無特別的發現


```python
df_orders['LevelThreeCategoryName'].unique()
```

```python
df_test =df_orders.groupby(['MemberId', 'LevelThreeCategoryName'])['Quantity'].sum().unstack(fill_value=0).astype('bool')
df_apriori = apriori(df_test, min_support=0.07, use_colnames=True)
df_conf = association_rules(df_apriori, metric="lift", min_threshold=1)
df_conf=df_conf[df_conf["confidence"]>0.6]
df_conf
```

## 顧客對特定商品類別購買關聯性 

貼身衣物的情況 其中舒適/日系這對組合較重要

```python
df_underwear = df_orders.groupby([ 'LevelTwoCategoryName']).get_group('內衣')
```

```python
df_test = df_underwear.groupby(['MemberId', 'LevelFourCategoryName'])['Quantity'].sum().unstack(fill_value=0).astype('bool')
df_apriori = apriori(df_test, min_support=0.01, use_colnames=True)
df_conf = association_rules(df_apriori, metric="lift", min_threshold=1)
df_conf
```



配件的情況 交易筆數較衣物類少很多


```python
df_accessory = df_orders.groupby([ 'LevelOneCategoryName']).get_group('鞋、包、配飾')
```

```python
df_orders['LevelOneCategoryName'].value_counts()
```



配件之間沒有值得一提的關聯性


```python
df_test = df_accessory.groupby(['MemberId', 'LevelFourCategoryName'])['Quantity'].sum().unstack(fill_value=0).astype('bool')
df_apriori = apriori(df_test, min_support=0.0001, use_colnames=True)
df_conf = association_rules(df_apriori, metric="lift", min_threshold=1)
df_conf
```

## 會員在使用優惠搭配的偏好

```python
df_mix = df_orders.merge(df_promotions, on='SalesOrderSlaveId', how='left')
df_mix.head()
```

```python
df_test = df_mix.groupby(['MemberId', 'DiscountType'])['Quantity'].sum().unstack(fill_value=0).astype('bool')
df_test.head()
```

可以發現在滿額折現的部分，會被最常拿來和其他優惠搭配

```python
df_apriori = apriori(df_test, min_support=0.01, use_colnames=True)
df_conf = association_rules(df_apriori, metric="lift", min_threshold=1)
df_conf.sort_values(by='antecedent support', inplace=True, ascending=False)
df_conf
```
