```python
import pandas as pd
import matplotlib as mp
import numpy as np
from apyori import apriori
from io import BytesIO as StringIO
```

```python
orders = pd.read_csv('../../Ntu_Orders.csv')
promotion_conditions = pd.read_csv('../../PromotionConditions.csv')
```

```python
orders.loc[:, ['TradesOrderGroupCode', 'LevelThreeCategoryName']].groupby(by='TradesOrderGroupCode')
```

```python
up = -1
orders_by_member = orders.loc[:,['LevelFourCategoryName','MemberId']].groupby('MemberId').groups
def construct(df_group, referenced_df, target):
    out = {}
    for it in df_group:
        out[it] = []
        for idx in df_group[it]:
            out[it].append(referenced_df.loc[idx, target])
    #return item_dict[row[groupby]]
    return out

item_dict = construct(orders_by_member, orders, 'LevelFourCategoryName')
item_dict
```

```python
relations = []
def convert(lst_dict, in_place = True):
    out = []
    for it in item_dict:
        if in_place:
            out.append(item_dict[it])
    return out
relations = convert(item_dict)
relations
```

```python
def clear_nan(transactions):
    for it in transactions:
        i = 0
        while(i < it.__len__()):
            if type(1.0) == type(it[i]) and np.isnan(it[i]):
                it.remove(it[i])
                continue
            else:
                i += 1
    return transactions

clear_nan(relations)
```

```python
results = apriori(relations, min_lift = 1.0)
results
```

```python
def filt(apriori_res, cnt = list(range(2,3)), min_prob = 0):
    tmp = []
    for it in apriori_res:
        if it.items.__len__() in cnt:
            try:
                if it.support >= min_prob:
                    tmp.append(it)
            except AttributeError as e:
                if it.confidence >= min_prob:
                    tmp.append(it)
    return tmp
filt(results, min_prob = 0.1, cnt=[2, 3, 4])
```
